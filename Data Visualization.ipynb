```python
import pandas

url ="https://raw.githubusercontent.com/sohamglobal/datasets/refs/heads/main/FitnessCustomers.csv"
df = pandas.read_csv(url)
df.head()
```

```python
df1=df[['CustomerID','BMI']].head(5)
print(df1)
```

```python
import seaborn
from matplotlib import pyplot
```

```python
seaborn.lineplot(x='CustomerID',y='BMI',data=df1)
pyplot.title("BMI Variation")
pyplot.show
```

```python
seaborn.barplot(x='CustomerID',y='BMI',data=df1)
pyplot.title("BMI Variation")
pyplot.show
```

```python
df2=df[['CustomerID','ExerciseTime']].head(5)
print(df2)
```

```python
seaborn.barplot(y='CustomerID',x='ExerciseTime',data=df2)
pyplot.title("Exercise Time")
pyplot.show
```

```python
seaborn.histplot(df['Gender'])
pyplot.title('Gender')
pyplot.show()
```

```python
df3 = df[['CustomerID','FitnessGoal','Age','Weight']].head(5)
```

```python
seaborn.barplot(x='CustomerID',y='Age',data=df3)
pyplot.title("Fitness Goals")
pyplot.show
```

```python
seaborn.histplot(df3['FitnessGoal'])
pyplot.title('Fitness Goals')
pyplot.show()
```

```python
seaborn.scatterplot(x='CustomerID',y='Age',data=df3)
pyplot.title('Age Distribution')
pyplot.show
```
